```python
import os
import cv2
from ultralytics import YOLO
from tqdm import tqdm

# ====== CONFIG ======
main_input_dir = r"D:\idd_temporal_test_2"  # your main folder
output_root = None # set to None if you don't want to save
model_path = r"D:\NLPInsuranceProject\AutonomousDrivingGanMV\idd detection\shareYolo\train\exp12\weights\best.pt"
img_extensions = ('.png', '.jpg', '.jpeg')
# =====================

# Load YOLOv8 model
model = YOLO(model_path)

# Walk through all subdirectories
for root, dirs, files in os.walk(main_input_dir):
    image_files = sorted([f for f in files if f.lower().endswith(img_extensions)])
    
    if not image_files:
        continue  # skip folders without images

    print(f"\n[INFO] Processing folder: {root}")

    # Mirror output structure if saving
    save_enabled = output_root is not None
    if save_enabled:
        rel_path = os.path.relpath(root, main_input_dir)
        output_folder = os.path.join(output_root, rel_path)
        os.makedirs(output_folder, exist_ok=True)

    for image_name in tqdm(image_files, desc=f"Detecting in {os.path.basename(root)}"):
        image_path = os.path.join(root, image_name)
        image = cv2.imread(image_path)

        # YOLO inference
        results = model.predict(source=image, imgsz=640, conf=0.3)

        # Annotate
        annotated = results[0].plot()

        # Show live detection
        cv2.imshow("YOLOv8 Detection", annotated)

        # Save annotated image
        if save_enabled:
            save_path = os.path.join(output_folder, image_name)
            cv2.imwrite(save_path, annotated)

        # Press 'q' to quit early
        if cv2.waitKey(1) & 0xFF == ord('q'):
            print("Stopped by user.")
            exit()

cv2.destroyAllWindows()
print("[DONE] All images processed.")
```

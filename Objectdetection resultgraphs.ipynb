```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your CSV
df = pd.read_csv(r'C:\Users\DELL\Documents\GitHub\AutonomousDrivingGanMV\objectDetectionresults.csv')

# Calculate F1-Score
df['F1_Score'] = 2 * (df['metrics/precision(B)'] * df['metrics/recall(B)']) / (df['metrics/precision(B)'] + df['metrics/recall(B)'])

```

### Using Matplotlib

```python


# Create figure and axes
fig, axs = plt.subplots(3, 2, figsize=(15, 12))
axs = axs.flatten()

# Plot Precision
axs[0].plot(df['epoch'], df['metrics/precision(B)'], marker='o', color='blue')
axs[0].set_title('Precision vs Epoch')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Precision')

# Plot Recall
axs[1].plot(df['epoch'], df['metrics/recall(B)'], marker='o', color='green')
axs[1].set_title('Recall vs Epoch')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Recall')

# Plot F1-Score
axs[2].plot(df['epoch'], df['F1_Score'], marker='o', color='red')
axs[2].set_title('F1-Score vs Epoch')
axs[2].set_xlabel('Epoch')
axs[2].set_ylabel('F1-Score')

# Plot mAP50
axs[3].plot(df['epoch'], df['metrics/mAP50(B)'], marker='o', color='purple')
axs[3].set_title('mAP50 vs Epoch')
axs[3].set_xlabel('Epoch')
axs[3].set_ylabel('mAP@0.5')

# Plot mAP50-95
axs[4].plot(df['epoch'], df['metrics/mAP50-95(B)'], marker='o', color='orange')
axs[4].set_title('mAP50-95 vs Epoch')
axs[4].set_xlabel('Epoch')
axs[4].set_ylabel('mAP@0.5:0.95')

# Hide the last empty subplot if any
fig.delaxes(axs[5])

# Tight layout
plt.tight_layout()
plt.show()
```
